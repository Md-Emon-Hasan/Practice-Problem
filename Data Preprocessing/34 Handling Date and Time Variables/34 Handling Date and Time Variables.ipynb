# Working with Dates

```python
# converting to datetime datetype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

__1.Extract Year__

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

__2.Extract Month__

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

#### Extract Days

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['days_of_week'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week name
date['day_name'] = date['date'].dt.day_name()
```

```python
date.head()
```

```python
# is weekend?
date['date_is_weekend'] = np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date.head()
```

#### Extract week of the year

```python
date['week_number'] = date['date'].dt.isocalendar().week
date.head()
```

#### Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)

date.head()
```

#### Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

np.round((today - date['date']) / np.timedelta64(1,'M'),0)
```
